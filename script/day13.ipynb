```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
# Load datasets for consumption and production

df_consumption = pd.read_csv("../data/coffee/coffeee/Coffee_domestic_consumption.csv")
df_production = pd.read_csv("../data/coffee/coffeee/Coffee_production.csv")
```

```python
df_consumption.head(5)
```

```python
df_production.head(5)
```

```python
# Join columns
df_joined = df_production.merge(df_consumption, on='Country', how='left')
df_joined = df_joined[["Country","Total_domestic_consumption","Total_production"]]
df_joined.head(5)
```

```python
# Clean country names
df_joined.replace("Bolivia (Plurinational State of)", "Bolivia", inplace=True)
df_joined.replace("Viet Nam", "Vietnam", inplace=True)
df_joined.replace("Lao People's Democratic Republic","Laos", inplace=True)
df_joined.replace("Trinidad & Tobago", "Bolivia", inplace=True)
df_joined.replace("Democratic Republic of Congo", "Congo DRC", inplace=True)
```

```python
# Read in geojson
world = gpd.read_file('../data/World_Countries_Generalized.geojson')
world.head(10)
```

```python
# Clean up country names somehow
world_coffee = world.merge(df_joined,left_on='COUNTRY', right_on='Country',how='left')
```

```python
# Export 
world_coffee.to_file("../data/coffee/coffee.geojson", driver="GeoJSON")  
```
